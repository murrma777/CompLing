Шаг 1: Установка библиотек

```python
# Устанавливаем библиотеки NLTK
import nltk
nltk.download('punkt')

# Устанавливаем библиотеку SpaCy
import spacy
nlp = spacy.load('en_core_web_sm')
```

Шаг 2: Загрузка текста

```python
#Загружаем текст через wget

!wget -O data.txt https://raw.githubusercontent.com/vifirsanova/compling/refs/heads/main/tasks/task1/data.txt
```

```python
# Читаем текст из файла
with open("data.txt", "r", encoding="utf-8") as file:
    data = file.read()
```

Шаг 3: Чистка текста

```python
# Удаляем пунктуацию

import re

cleaned_text = re.sub(r'[^\w\s]', '', data) # \w обозначает любой символ слова (буквы, цифры);
# \s это пробелы и перенос строки; ^ означает "не".
# Так мы убираем все символы, кроме перечисленных (буквы и пробелы)

print("Текст без пунктуации:", cleaned_text)
```

Шаг 4: Приведение текста к нижнему регистру

```python
# Приводим к нижнему регистру
cleaned_text = cleaned_text.lower()
print("Текст в нижнем регистре:", cleaned_text)
```

Шаг 5: Удаление стоп-слов

```python
# Импортируем библиотечки

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# Загружаем список стоп-слов
nltk.download('stopwords')

# Удаляем стоп-слова с помощью NLTK
stop_words = set(stopwords.words('russian'))  # Загрузка списка стоп-слов для русского языка

# Токенизация текста
words = word_tokenize(cleaned_text)

# Фильтруем слова
filtered_words = [word for word in words if word not in stop_words]

# Выводим результат
print("Текст без стоп-слов:", ' '.join(filtered_words))
```

Шаг 6: Лемматизация

```python
# Код для лемматизации с помощью SpaCy
doc = nlp(cleaned_text)

# Лемматизация слов
lemmatized_words = [token.lemma_ for token in doc]

# Выводим результат
print("Лемматизированный текст:", ' '.join(lemmatized_words))
```

Шаг 7: Вывод результатов

```python
# Выводим переменную с текстом

print("Итоговый текст:", cleaned_text)
```
